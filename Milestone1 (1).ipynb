```python
import pandas as pd

df = pd.read_csv("train.txt", sep="\t", names=["head", "relation", "tail"])

df.head()
```

```python
print("Shape of dataset:", df.shape)
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df = df.reset_index(drop=True)
```

```python
print("Unique Heads:", df["head"].nunique())
print("Unique Relations:", df["relation"].nunique())
print("Unique Tails:", df["tail"].nunique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.countplot(y="relation", data=df, order=df["relation"].value_counts().index[:10])
plt.title("Top 10 Relations in Knowledge Graph")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le_head = LabelEncoder()
le_relation = LabelEncoder()
le_tail = LabelEncoder()

df["head_enc"] = le_head.fit_transform(df["head"])
df["relation_enc"] = le_relation.fit_transform(df["relation"])
df["tail_enc"] = le_tail.fit_transform(df["tail"])

df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Normalize ONLY head and tail
df[["head_enc","tail_enc"]] = scaler.fit_transform(
    df[["head_enc","tail_enc"]]
)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X = df[["head_enc","tail_enc"]]
y = df["relation_enc"]   # keep original encoded values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

pred = model.predict(X_test)

print("Model Accuracy:", accuracy_score(y_test, pred))
```
